```python
import spacy
from tqdm import tqdm

de_parser = spacy.load("de_core_news_sm")

def shallow_dependencies2(sents, parser, roles=None):
    roles = roles if roles is not None else['nsubj', 'verb', 'dobj', 'pobj']
    docs = parser.pipe(sents)
    relations = []
    for doc in docs:
        realizations = {r: ['', []] for r in roles}
        for i, tok in enumerate(doc):
            for r in roles:
                if r == 'verb':
                    if tok.dep_ == 'ROOT' and tok.pos_ == 'VERB':
                        realizations[r] = [tok.text, [tok.i]]
                else:
                    if tok.dep_ == r and realizations[r][0] == '':
                        realizations[r] = [' '.join([toki.text for toki in tok.subtree]),
                                           [toki.i for toki in tok.subtree]]

        relations.append({'text':{k: v[0] for k, v in realizations.items()},
                         'idx': {k: v[1] for k, v in realizations.items()}})
    return relations
rl = shallow_dependencies2(['johnson drohte sich zurückzuziehen..',
                            'goistering bezieht sich auf das lachen eines babys..'], de_parser, roles=['sb', 'oa', 'da',
                                                                                                       'og', 'op'])

roles = ['sb', 'verb', 'oa', 'da', 'og', 'op', 'oa2', 'oc']
orig_sens = []
rls = []
for i, training_batch in tqdm(enumerate(data.train_iter)):
    sents = [' '.join([data.vocab.itos[t] for t in text_i]).replace('<go> ', '').replace(' <pad>', '').replace(' <eos>', '') 
           for text_i in training_batch.text[:]]
    rl = shallow_dependencies2(sents, de_parser, roles=roles)
    rls.extend(rl)
    orig_sens.extend(sents)
rlts = [rl['text'] for rl in rls]
```

```python
sr_number = {r: 0 for r in roles}
for rlt in rlts:
    for k, v in rlt.items():
        if v!='':
            sr_number[k]+=1
            
```
